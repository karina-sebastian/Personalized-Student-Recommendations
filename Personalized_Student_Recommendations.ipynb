```python


# Extract title and correct_answer_marks
titles = []
correct_answer_marks_list = []

for index, row in historical_quiz_df.iterrows():
    try:
        quiz_dict = row['quiz']
        titles.append(quiz_dict.get('title', 'N/A'))  # Use .get() to handle missing keys
        correct_answer_marks_list.append(quiz_dict.get('correct_answer_marks', 'N/A'))
    except (KeyError, TypeError, AttributeError):
        titles.append('N/A')
        correct_answer_marks_list.append('N/A')

# Add the extracted data as new columns to the DataFrame
historical_quiz_df['title'] = titles
historical_quiz_df['correct_answer_marks'] = correct_answer_marks_list

# Display the updated DataFrame
print(historical_quiz_df[['title', 'correct_answer_marks']])
```

```python
def generate_recommendations(user_id):
    # Accessing the global variables
    global weak_topics
    global weak_difficulty

    # weak_topics and weak_difficulty are already lists, so no need for tolist()
    weak_topics_list = [str(topic) for topic in weak_topics.index]  # Get weak topic titles
    weak_difficulty_list = [str(diff_level) for diff_level in weak_difficulty.index] # Convert all elements to strings

    recommendations = []

    if weak_topics_list: # check for empty list
        recommendations.append(f"Focus on the following weak topics: {', '.join(weak_topics_list)}.")
    else:
        recommendations.append("Keep up the good work on the topics!")

    if weak_difficulty_list: # check for empty list
        recommendations.append(f"Try to focus on questions with difficulty levels: {', '.join(weak_difficulty_list)}.")
    else:
        recommendations.append("You seem to be handling all difficulty levels well!")

    return recommendations, weak_topics_list # Return recommendations and weak topic titles
```
